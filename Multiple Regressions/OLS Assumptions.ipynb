# OLS Assumptions

# 1 Linearity

The specified model must represent a linear relationship. The graph has a straight line.



 Assumption is violated:
- we see no straight line



 FIXES:
1. Run a non-linear regression
2. Exponential transformation
3. Log Transformation

# 2. No Edogeneity

The independent variables shouldn't be correlated with the error term.

When independent variables correlated with error term, it is called "Omitted Variable Bias".

"`Omitted Variable Bias`" - you forget to include a relevant variable.

Everything you don't explain with the model explain to the error.



When in doubt, just include the variables.



# 3. Normality and Homoscedasticity



What we should if it is NOT _normally_ distributed?
- Central Limit Theory applies

If the _mean_ is not expected to be ZERO, the line is NOT the best-fitting line. But in real life, this is rare. Having an intercept solves the problem.

`Homoscedasticity` - means EQUAL variance



 Prevention for Heteroscedasticity:
1. Look for OVB
2. Look for outliers and try to remove them
3. Transformation (Log)
   - Take the log of the variable
   - Perform the regression

# 4. NO AUTOCORRELATION

No identifiable relationship should exist between the values of the error term.

Autocorrelation is very highly common in "time series data" i.e. stock prices.

 Detection:
- If there are no patterns to be seen => no autocorrelation
- Durbin-Watson
  

 NO REMEDY

"When in the presences of autocorrelation avoid the linear regression model."

REGRESSION FOR TIME SERIES DATA
- Autoregressive model
- Moving Average Model
- Autoregressive Moving Average Model
- Autoregressive Integrated Moving Average Model

# 5. NO MULTICOLLINEARITY

No predictor variable should be perfectly (or almost perfectly explained by the other predictors)



Perfect multicollinearity imposes a big problem on our model.



Same also with the imperfect multicollinearity

Real life examples:


 FIXES:
1. Drop one of the 2 variables
2. Transform them into one (e.g. average price)
3. Keep them both (but be cautious )

Prevention:
- Find the correlation between each 2 pairs of independent variables
