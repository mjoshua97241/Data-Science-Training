# Test For Significance of the Model (F-Test)

F-statistic is used for testing the overall significance of the model

F-test:

$H_{0}$: $ß_{1}$ = $ß_{2}$ = ... = $ß_{k}$ = 0

$H_{1}$: at least one $ß_{i}$ ≠ 0

Interpretation: If all betas are 0, then none of the Xs matter => our model has no merit
