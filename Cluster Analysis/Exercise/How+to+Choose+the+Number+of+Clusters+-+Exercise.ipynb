## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries_exercise.csv'</i>.

```python
data = pd.read_csv('Countries_exercise.csv')
data.head()
```

Remove the duplicate index column from the dataset. 

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(data['Longitude'], data['Latitude'], color = 'C0')

plt.xlabel('Longitude', fontsize = 20)
plt.ylabel('Latitude', fontsize = 20)
plt.show()
```

## Select the features

Make sure to select the appropriate features since we are no longer using the categorical variable for our clustering but rather <i>'Longitude'</i> and <i>'Laditude'</i>.

```python
x = data.iloc[:, 1:]
x.head()
```

## Clustering

Use 4 clusters initially.

```python
kmeans = KMeans(4)
```

```python
kmeans.fit(x)
```

### Clustering results

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters.head()
```

Plot the data once again but separate the data by the clusters we defined.  

## Plot the data

```python
plt.scatter(data_with_clusters['Longitude'], data_with_clusters['Latitude'], c=data_with_clusters['Cluster'], cmap='rainbow')

plt.xlabel('Longitude', fontsize = 20)
plt.ylabel('Latitude', fontsize = 20)
plt.show()
```

## Selecting the number of clusters

### WCSS (within-cluster sum of squares)

Use the ingerated <i>sklearn</i> method <i> 'inertia_' </i>.

```python
# Get the WCSS for the current solution
kmeans.inertia_
```

Write a loop that calculates and saves the WCSS for any number of clusters from 1 up to 10 (or more if you wish).

```python
# Create an empty list
wcss = []

# Create all possible cluster solution with a loop
for n_clusters in range (1,10):
    # Cluster solution with i clusters
    kmeans = KMeans(n_clusters)
    # Fit the data
    kmeans.fit(x)
    # Find WCSS for the current iteration
    wcss_iter = kmeans.inertia_
    # Append the value to WCSS list
    wcss.append(wcss_iter)
```

```python
wcss
```

### The Elbow Method

```python
# Create a variable containing the numbers from 1 to 10, so we can use it as X axis of the future plot
number_clusters = range(1,10)

# Plot the number of clusters vs WCSS
plt.plot(number_clusters, wcss)

# Name the graph
plt.title('The Elbow Method')

# Name the x-axis
plt.xlabel('Number of clusters(k)')

# Name the y-axis
plt.ylabel('Within-cluster Sum of Squares')
plt.show()
```
